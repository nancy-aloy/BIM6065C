(global variables _vs_ local variables)

The functions that you define typically build on each others, each time allowing you to forget their inner working. As much as possible try to only refer to variables that are passed (here: s and nuc) or local (here: count, c), relying on global variables means you need to remember to set them properly.

It is a good idea to put efforts in making your own code simpler to understand. Never hesitate to rewrite, reorganize, adjust variables and function names. It feels as wasted time (it doesn't do anything new!), but it is necessary to avoid getting "stuck".

```python
ATGnewseq = "CAGTGTGCATAGATGTGTA"
```

```python
def posATG(my_seq):   
    st = 0
    pos = 0
    for c in my_seq:
        if c == "A" and st == 0:
            st = 1
        elif c == "T" and st == 1:
            st = 2
        elif c == "G" and st == 2:
            break
        else:
            st = 0
        pos = pos + 1
        pos_ATG = pos - 2
    print(pos_ATG)

posATG(ATGnewseq)
posATG(seq)
posATG(ATGseq)
```

### 3) Where is the first stop "codon"? (TAA, TAG, TGA) *Make it into a...*

What if you had a more versatile version of 2?

(min)

```python
TAA = "AGCTATGCTGCATAAGCTGCATG"
```

```python
def find_stop_codons(sequence):
    st = 0
    pos = 0
    while pos < len(sequence):
        c = sequence[pos]
        if c == "T" and st == 0:
            st = 1
        elif c == "A" and st == 1:
            st = 2
        elif c == "A" and st == 2:
            print("TAA", pos - 2)
            st = 0
        elif c == "G" and st == 2:
            print("TAG", pos - 2)
            st = 0
        elif c == "G" and st == 1:
            st = 2
        elif c == "A" and st == 2:
            print("TGA", pos - 2)
            st = 0
        else:
            st = 0
        pos = pos + 1

find_stop_codons(ATGnewseq)
find_stop_codons(TAA)
```

### 4) How many CpG? *Mak...*

```python
def count_CpG(sequence):
    count = 0
    pos = 0
    while pos < len(sequence) - 1:
        if sequence[pos] == "C" and sequence[pos + 1] == "G":
            count = count + 1
        pos = pos + 1

    return count

count_CpG(ATGseq)
```

### 5) How many 'ATG' are present in sequences from the file 'seq_100.txt'?

These were loaded in the `mseq` variable at the beginning of the class. They come from the file `seq_100.txt` (on Studium).

```python
def count_ATG(sequence):
    count = 0
    pos = 0
    while pos < len(sequence) - 2: 
        if sequence[pos] == "A" and sequence[pos + 1] == "T" and sequence[pos + 2] == "G":
            count += 1
        pos += 1

    return count

ATGseq = "CGATGCTATGCGTAGAT"
count_ATG(ATGseq)
list_mseq = [x.strip() for x in mseq]
long_seq_mseq = ''.join(list_mseq)
count_ATG(long_seq_mseq)
```
